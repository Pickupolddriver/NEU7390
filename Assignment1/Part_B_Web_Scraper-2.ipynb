```python
http = urllib3.PoolManager()
import csv
import warnings
warnings.filterwarnings('ignore')
result=[]
#create a result list;
import datetime
for url in l:
    #find links in <a>;
    tmp=[]
    http = urllib3.PoolManager()
    if re.match(r'^https?:/{2}\w.+$', url['href']):
        
        #Regular Expression, the link will begin with http[s]
      uuurl=url['href']#find the href in url-- <a>
      tmp.append(uuurl)
        #link 
      tmp.append(url.text)
    #text
      tmp.append(datetime.datetime.now().strftime("%A %B %-d, %Y %I:%M %p"))
        #last check
      response = http.request('GET',uuurl)
    # Test whether the link is accessble
      if response.status ==200 :
            tmp.append("TRUE")
            print (tmp)
      else :
          #tmp.append(url['href'])
          #tmp.append(url.text)
          #tmp.append(datetime.datetime.now().strftime("%A %B %-d, %Y %I:%M %p"))
          tmp.append("FALSE")
          print (tmp)
      result.append(tmp)
    #print (tmp)
    
    
    #links.append(tmp)
```
